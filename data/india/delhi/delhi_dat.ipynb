```python
#%pip install jpype1 tabula-py
import tabula
import pandas as pd
```

```python
def extract_tables_from_pdf(pdf_path, pages='all'):
    try:
        tables = tabula.read_pdf(
            pdf_path, 
            pages=pages, 
            multiple_tables=True
        )
        
        cleaned_tables = []
        for i, table in enumerate(tables):
            if i == 0:
                continue
            table.columns = ['AC_NO', 'Constituency', 'Candidate', 'Address', 
                           'Phone', 'Email', 'Party', 'Status']
            cleaned_tables.append(table)
        
        final_df = pd.concat(cleaned_tables, ignore_index=True)
        final_df = final_df.iloc[1:]  # Drop first row
        final_df = final_df.replace('\r', ' ', regex=True)
        final_df = final_df.dropna(how='all')
        final_df = final_df.reset_index(drop=True)
        final_df.to_csv('assembly_members.csv', index=False)
        return final_df

    except Exception as e:
        print(f"Error extracting tables: {str(e)}")
        return None
```

```python
pdf_path = "../../pwned_pols/data/india/delhi_7th_assembly.pdf"
df = extract_tables_from_pdf(pdf_path)
```
