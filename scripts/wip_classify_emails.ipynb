```python
import pandas as pd 
pd.set_option('display.max_rows', 300)
```

```python
df = (
    pd.read_csv("../data/email_lvl_cov.csv")
    .assign(domain=lambda df_: df_["email"].str.split("@").str[1])
)
df
```

```python
(
    df
    .query("ecategory=='Other'")
    .groupby("domain")
    .size()
    .reset_index(name="count")
    .sort_values("count", ascending=False, ignore_index=True)
)
```

```python
(
    df
    .query("ecategory=='Official'")
    .groupby("domain")
    .size()
    .reset_index(name="count")
    .sort_values("count", ascending=False, ignore_index=True)
)
```

```python
(
    df
    .query("ecategory=='Commercial'")
    .groupby("domain")
    .size()
    .reset_index(name="count")
    .sort_values("count", ascending=False, ignore_index=True)
)
```
